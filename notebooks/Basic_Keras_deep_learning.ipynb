# In this notebook we create some data and train a simple network

```python
import numpy as np
from random import randint
from sklearn.preprocessing import MinMaxScaler
```

```python
#Creating empty lists
train_samples=[]
train_labels=[]
```

```python
#Generating the Data
for i in range(50):
    random_younger=randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
    
    random_older=randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)
    
for i in range(1000):
    random_younger=randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)
    
    
    random_older=randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)
```

```python
for i in train_samples:
    print (i)
```

```python
for i in train_labels:
     print(i)
```

```python
train_labels=np.array(train_labels)
train_samples=np.array(train_samples)
```

```python
train_labels.dtype
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform((train_samples).reshape(-1,1))
```

```python
for i in scaled_train_samples:
    print(i)
```

```python
#Creating an artificial neural network
import keras
import numpy as np
from keras import backend as k
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
model = Sequential([Dense(16,input_shape=(1,),activation='relu'),
                   Dense(32,activation='relu'),Dense(2,activation='softmax')])
```

```python
model.summary()
```

```python
#Train an articial neural network
model.compile(Adam(lr=0.001), loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
#Build a validation set
model.fit(scaled_train_samples,train_labels,validation_split=0.1,batch_size=20,epochs=20,shuffle=True,verbose=1)
```

# Preprocess the test data

```python
test_labels=[]
test_samples=[]
```

```python

for i in range(50):
    #the 5% of younger people who got experiences side effects
    random_younger=randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(1)
    
    #the 95% of old people who got experience side effects
    random_younger=randint(65,100)
    test_samples.append(random_younger)
    test_labels.append(0)
    
for i in range(200):
    #the 95% of younger people who got experienced side effects
    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(0)
    
    #the 5% of old people who got experienced the side effects
    random_younger = randint(65,100)
    test_samples.append(random_younger)
    test_labels.append(1)
```

# Confusion matrix


```python

from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
%matplotlib  inline
```

```python
cm = confusion_matrix(test_labels,rounded_predictions)
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
cm_plot_labels = ['no side effects','sid effects']
plot_confusion_matrix(cm,cm_plot_labels,title ='Confusion Matrix')
```
